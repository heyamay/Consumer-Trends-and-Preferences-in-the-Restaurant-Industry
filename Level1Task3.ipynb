```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\Dongen_Master\Desktop\L3T3\Dataset .csv"  # Update with your file path
data = pd.read_csv(file_path)
```

```python
print(data.head())
print(data.columns)
```

```python
# Calculate the count of restaurants in each price range
price_range_counts = data['Price range'].value_counts()

# Calculate the percentage of restaurants in each price range
price_range_percentages = (price_range_counts / price_range_counts.sum()) * 100

# Display the counts and percentages
print("Price Range Counts:")
print(price_range_counts)
print("\nPrice Range Percentages:")
print(price_range_percentages)
```

```python
# Create a bar chart for the price range distribution
plt.figure(figsize=(10, 6))
price_range_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the figure
plt.savefig('price_range_distribution.png')  # Save as PNG file

# Show the figure
plt.show()
```
