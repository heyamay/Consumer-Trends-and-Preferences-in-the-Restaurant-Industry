```python
import pandas as pd

# Load the dataset using the correct file path
file_path = r"C:\Users\Dongen_Master\Desktop\1_Internship\Dataset .csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to confirm it loaded correctly
print(df.head())
```

```python
# Assuming the column name is 'cuisines'
cuisines_counts = df['Cuisines'].value_counts()
```

```python
top_three_cuisines = cuisines_counts.head(3)
print(top_three_cuisines)
```

```python
total_restaurants = len(df)

# Calculate percentage
percentage_top_cuisines = (top_three_cuisines / total_restaurants) * 100
print(percentage_top_cuisines)
```

```python
# Combine the counts and percentages into a DataFrame
results = pd.DataFrame({
    'Cuisines': top_three_cuisines.index,
    'Count': top_three_cuisines.values,
    'Percentage': percentage_top_cuisines.values
})

print(results)

# Optionally, save the results to a CSV file
results.to_csv('top_cuisines_results.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data for visualization
cuisines = ['North Indian', 'North Indian, Chinese', 'Chinese']
counts = [936, 511, 354]
percentages = [9.80002094, 5.350225107, 3.706418176]
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(cuisines, counts, color='skyblue')

# Add title and labels
plt.title('Top 3 Cuisines by Count')
plt.xlabel('Cuisines')
plt.ylabel('Count of Restaurants')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(percentages, labels=cuisines, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'gold', 'lightskyblue'])

# Add title
plt.title('Percentage of Restaurants Serving Top Cuisines')

# Show the plot
plt.show()
```
